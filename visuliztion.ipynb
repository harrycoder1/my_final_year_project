```python
df_week = pd.read_csv('weekly_sum.csv')
df_week.drop("Unnamed: 0" , axis=1 , inplace=True)
df_week.head(5)
```

```python
plt.figure(figsize=(10,10))
# sns.lineplot(y='litreWater' , x='start_time' , data=df_week )
plt.xticks(rotation=45)  
plt.tight_layout()  
plt.xticks(fontsize=8)  
plt.xlabel("weeks")
plt.ylabel("litre Water")
plt.legend()
plt.title("Week Uses vs Litre_water")
plt.plot(df_week['start_time'], df_week['litreWater'] , lw=2 , marker="." , )
# plt.legend()

plt.show()
```
