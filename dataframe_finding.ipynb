```python

df['start_time'] = pd.to_datetime(df['start_time'] )

# Make a copy of the DataFrame
df1 = df.copy()

# Set 'start_time' as the index
df1.set_index('start_time', inplace=True)


def get_data_by_filter ( date_flag , value):
    return df1.resample(date_flag)[value ] 

# monthly_duration = df1.resample('M')['duration' ] 

```

#### Visulizing on the monthly basis

```python
monthly_duration =  get_data_by_filter('M' , 'duration')
monthly_litre_water =  get_data_by_filter('M' , 'litreWater')
# print(type(monthly_duration))
print("Total Duration of motor run")

m_d = monthly_duration.sum().reset_index('start_time')
# print(m_d)

# print("Total Litre of Water has used :")
m_l = monthly_litre_water.sum().reset_index('start_time')
# print(m_l)

new_data_set_monthly = pd.merge(m_d , m_l  , on='start_time'  , how="inner")
# new_data_set_monthly = pd.concat([monthly_duration , monthly_duration] , axis=1)
new_data_set_monthly.to_csv("monthly_sum.csv")

new_data_set_monthly
```

### data on weekly basis 

```python
weekly_duration = get_data_by_filter('W' , 'duration')
weekly_water_litre = get_data_by_filter('W' , 'litreWater')

w_d = weekly_duration.sum().reset_index('start_time')
w_l = weekly_water_litre.sum().reset_index('start_time')


new_weekly_data = pd.merge(w_d, w_l , on="start_time" , how="inner")
new_weekly_data.to_csv('weekly_sum.csv')

new_weekly_data.head(10)
```

### Data on Daily basis 

```python
daily_duration = get_data_by_filter('D' , 'duration')
daily_litre_water = get_data_by_filter('D' , 'litreWater')

d_d = daily_duration.sum().reset_index('start_time')
d_l = daily_litre_water.sum().reset_index('start_time')

new_data_set_daily = pd.merge(d_d , d_l , on="start_time" , how="inner")
new_data_set_daily.to_csv('daily_sum.csv')
new_data_set_daily.head(10)
```
